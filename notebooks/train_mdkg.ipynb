# Training BioBERT + SpERT on MDKG
This notebook trains a biomedical relation extraction model on the Mental Disorder Knowledge Graph (MDKG) dataset.

```python
import os, sys

os.makedirs("../logs/mdkg_biobert", exist_ok=True)
os.makedirs("../models/mdkg_biobert", exist_ok=True)
```

## Create Output Directories
Directories for logs and saved model checkpoints are created if they do not already exist.

```python
config_text = """label = mdkg_biobert
model_type = spert
model_path = dmis-lab/biobert-base-cased-v1.1
tokenizer_path = dmis-lab/biobert-base-cased-v1.1
train_path = ../data/spert_mdkg/train.json
valid_path = ../data/spert_mdkg/dev.json
types_path = ../data/spert_mdkg/types.json
train_batch_size = 1
eval_batch_size = 1
neg_entity_count = 100
neg_relation_count = 100
epochs = 5
lr = 5e-5
lr_warmup = 0.1
weight_decay = 0.01
max_grad_norm = 1.0
rel_filter_threshold = 0.4
size_embedding = 25
prop_drop = 0.1
max_span_size = 10
store_predictions = true
store_examples = true
sampling_processes = 1
max_pairs = 1000
final_eval = true
log_path = ../logs/mdkg_biobert
save_path = ../models/mdkg_biobert
"""

with open("../spert/configs/mdkg_train.conf", "w") as f:
    f.write(config_text)

print("MDKG config saved.")
```

## Define MDKG Training Configuration
Training hyperparameters and dataset paths are configured for MDKG experiments.
